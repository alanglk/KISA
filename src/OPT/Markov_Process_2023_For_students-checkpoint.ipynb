# Markov Process 

### Garden example

We start by reproducing the garden example analyzed in the lecture.   

In this example we have three states: (1) Good. (2) Regular. (3) Bad.  
    
We consider two possible actions: (1) not fertilize and  (2) fertilize   

The matrices of transition probabilities and reward probabilities for the two action  
are displayed below.   
    

```python
%%%latex

\[
    P1 =\left(
                \begin{array}{ccc}
                   0.2 & 0.5  & 0.3 \\
                   0 & 0.5 & 0.5 \\
                   0 & 0 & 1
        \end{array}
              \right).
  \] 

\[
    R1 =\left(
                \begin{array}{ccc}
                   7 & 6  & 3 \\
                   0 & 5 & 1 \\
                   0 & 0 & -1
        \end{array}
              \right).
  \]



\[
    P2 =\left(
                \begin{array}{ccc}
                   0.3 & 0.6  & 0.1 \\
                   0.1 & 0.6 & 0.3 \\
                   0.05 & 0.4 & 0.55
        \end{array}
              \right).
  \]

\[
    R2 =\left(
                \begin{array}{ccc}
                   6 & 5  & -1 \\
                   7 & 4 & 0 \\
                   6 & 3 & -2
        \end{array}
              \right).
  \]

```

### Garden problem. Finite horizon analysis.

We will use the mdptoolbox to define and solve the Markov Decision Process (MDP)  

Execute the following cell to initialize the MDP
of the Garden problem. 

```python
# P1: Transition matrix for action 1
# R1: Reward matrix for action 1

P1 = np.array([[0.2,0.5,0.3],[0,0.5,0.5],[0,0,1]])
R1 = np.array([[7,6,3],[0,5,1],[0,0,-1]])

# P2: Transition matrix for action 1
# R2: Reward matrix for action 1

P2 = np.array([[0.3,0.6,0.1],[0.1,0.6,0.3],[0.05,0.4,0.55]])
R2 = np.array([[6,5,-1],[7,4,0],[6,3,-2]])

# Array of matrices containing all transition and reward matrices
AllPs = np.array([P1,P2])
AllRs = np.array([R1,R2])
```

In the next step we use dynamic programming to find the optimal policy in a finite horizon scheme.  
Execute the following cell to initialize the algorithm.


```python
# We define the FiniteHorizon algorithm 
number_stages = 3
discount = 1
fh = mdptoolbox.mdp.FiniteHorizon(AllPs,AllRs, discount, number_stages)
```

Notice the warning in the definition of the algorithm. In the general case we can not assume its convergence.
Execute the following cell to execute the algorithm and visualize the results. 

```python
fh.run()               # We run the algorithm

Vs_per_Stages = fh.V   # Expected rewards by stages

BestPolicy = fh.policy

print("Expected rewards in each stage: ")
print(Vs_per_Stages)
print("Best policy :") 
print(BestPolicy)

```

In the printed results, take some time to compare the results of the expected rewards and best policy  produced by the algorithm  with those discussed in the lecture in the explanation of the dynamic programming model. 

Notice that the algorithm also outputs best policy for each of the three stages. Also notice that mdptoolbox assumes the enumeration of the actions start from 0. Therefore we have action 0 (not fertilize) and action 1 (fertilize).

Run the following cell to visualize the optimal mean reward for each state in the different stages

```python
x1 = np.array([0,1,2,3])
plt.plot(x1, Vs_per_Stages[0,:], label=r'Good')
plt.plot(x1, Vs_per_Stages[1,:], label=r'Regular')
plt.plot(x1, Vs_per_Stages[2,:], label=r'Poor')
plt.ylabel(r'v')
plt.xlabel(r'stages')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

### Garden problem. Policy iteration algorithm

In the next example we will use the algorithm of policy iteration to compute the best policy. There are two possible ways to solve the problem. Using a system of linear equations or using the iterative algorithm analyzed in the lecture. We use the second approach as discussed in the lecture. 


```python
# Using the arrays of transition matrices and rewards from the garden experiment, a discount factor alpha=0.1 and
# an iterative procedure to find the optimal policies, we apply the policy iteration  algorithm

pi = mdptoolbox.mdp.PolicyIteration(AllPs, AllRs, discount=1, eval_type="iterative")
pi.run()

Vs_per_Stages = pi.V   # Expected rewards for each state

BestPolicy = pi.policy  # Best policy

print("Optimal value functions: ")
print(Vs_per_Stages)
print("Best policy :") 
print(BestPolicy)
```

```python

# Display of the expected rewards
width = 0.35               # Width of the bars
nstates = 3                # Number of the states
ind = np.arange(nstates)   # x locations for the groups
fig, ax = plt.subplots()
rects1 = ax.bar(ind,Vs_per_Stages, width, color='r')
plt.ylabel(r'v')
plt.xlabel(r'states')
plt.show()
```
